```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
x = [[30, 40], [25, 80], [35, 30], [28, 75], [32, 65], [40, 20], [22, 85]]
y = ['Sunny', 'Rainy', 'Sunny', 'Rainy', 'Rainy', 'Sunny', 'Rainy']
q = [[29, 60]]
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x, y)
```

```python
predictions = knn.predict(q)
predictions
```

```python
for (x1, y1), label in zip (x, y):
    color = 'Yellow' if label == 'Sunny' else 'Blue'
    plt.scatter(x1, y1, c=color)
```

```python
qx, qy = q[0]
plt.scatter(qx, qy, color='Orange', marker='*', s=200)
plt.title(f"Test data Prediction={predictions}")
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.grid(True)
plt.show()
```
