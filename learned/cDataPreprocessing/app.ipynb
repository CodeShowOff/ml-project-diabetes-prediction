```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder, OneHotEncoder
```

```python
data = {'age': [27, 28, 29, None, 30, 31, 32, None, 29, 22],
        'height': [160, 162, 165, None, 167, None, 168, 169, None, 170],
        'wt': [56, 57, 58, 59, 60, None, 61, 62, None, 65],
        'city': ['kolkata', np.nan, 'up', 'bihar', 'kanpur', np.nan, 'china', 'purnea', np.nan, 'kolkata'],
        'marks': [60, 65, None, 70, 75, 78, 79, 80, 81, 82]
}
```

```python
df = pd.DataFrame(data)
df 
```

```python
print("Total missing_values:", df.isnull().sum())
```

```python
df_drop = df.dropna()
df_drop
```

```python
df_drop.to_csv('user.csv')
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
df
```

```python
df['height'].fillna(df['height'].median(), inplace=True)
df
```

```python
# df['wt'].fillna(method='ffill', inplace=True)
df['wt'].fillna(df['wt'].ffill(), inplace=True)
df
```

```python
df['city'].fillna(df['city'].mode()[0], inplace=True)
df
```

```python
df['marks'].fillna(method='bfill', inplace=True)
df
```

```python
mm = MinMaxScaler()
std = StandardScaler()
```

```python
df[['mm_age', 'mm_height']] = mm.fit_transform(df[['age', 'height']]) #normalization
df[['std_wt']] = std.fit_transform(df[['wt']]) #standardization
df
```

```python
le = LabelEncoder()
df['le_city'] = le.fit_transform(df['city'])
df
```

```python
ohe = OneHotEncoder(sparse_output=False)
one_array = ohe.fit_transform(df[['city']])
ohe_columns = ohe.get_feature_names_out(['city'])
df_ohe = pd.DataFrame(one_array, columns=ohe_columns)
df_ohe
```

```python
le2 = LabelEncoder()
df['le2_city'] = le2.fit_transform(df['city'])
df
```
