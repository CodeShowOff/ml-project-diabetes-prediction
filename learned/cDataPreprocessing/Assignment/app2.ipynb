```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
```

```python
data_emp = {
    'Emp_ID': [201, 202, 203, 204, 205, 206, 207],
    'Age': [23, 29, None, 35, 40, 45, 31],
    'Salary': [25000, 32000, 28000, None, 50000, 60000, 35000],
    'Experience': [1, 3, 2, 5, None, 8, 4],
    'Performance_Score': [3.5, None, 4.0, 4.5, 4.2, None, 3.8],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female']
}
```

```python
df = pd.DataFrame(data_emp)
df 
```

```python
print("Total missing_values:", df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

```python
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df
```

```python
df['Experience'].fillna(df['Experience'].mean(), inplace=True)
df
```

```python
df['Performance_Score'].fillna(df['Performance_Score'].median(), inplace=True)
df
```

```python
mm = MinMaxScaler()
std = StandardScaler()
le = LabelEncoder()
```

```python
df['le_Gender'] = le.fit_transform(df['Gender'])
df
```

```python
df[['mm_Age', 'mm_Salary']] = mm.fit_transform(df[['Age', 'Salary']]) #normalization
df[['std_Performance_Score']] = std.fit_transform(df[['Performance_Score']]) #standardization
df
```
