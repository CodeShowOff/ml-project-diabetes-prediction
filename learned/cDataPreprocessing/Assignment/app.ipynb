```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
```

```python
data = {
    'Patient_ID': [101, 102, 103, 104, 105, 106, 107],
    'Age': [25, 30, 28, None, 45, 50, 29],
    'Weight': [58, 75, None, 55, 85, 90, None],
    'BloodPressure': [120, None, 130, 110, 140, 150, None],
    'Cholesterol': [200, 210, 220, 190, None, 250, 230],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female']
}
```

```python
df = pd.DataFrame(data)
df 
```

```python
print("Total missing_values:", df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

```python
df['Weight'].fillna(df['Weight'].median(), inplace=True)
df
```

```python
df['BloodPressure'].fillna(df['BloodPressure'].mean(), inplace=True)
df
```

```python
df['Cholesterol'].fillna(df['Cholesterol'].median(), inplace=True)
df
```

```python
mm = MinMaxScaler()
std = StandardScaler()
le = LabelEncoder()
```

```python
df['le_Gender'] = le.fit_transform(df['Gender'])
df
```

```python
df[['mm_Age', 'mm_Weight']] = mm.fit_transform(df[['Age', 'Weight']]) #normalization
df[['std_Cholesterol']] = std.fit_transform(df[['Cholesterol']]) #standardization
df
```
