```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
data = {
    'x': [4, 6, 8, 9],
    'y': [10, 20, 30, 40]
}
```

```python
df = pd.DataFrame(data)
df 
```

```python
X = df - df.mean()
X
```

```python
cov_matrix = np.cov(X.T)
cov_matrix
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(X)
```

```python
eigen_values = pca.explained_variance_
eigen_values
```

```python
eigen_vectors = pca.components_
eigen_vectors
```
