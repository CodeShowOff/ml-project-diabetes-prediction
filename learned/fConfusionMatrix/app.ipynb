```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
y_act = [1, 1, 1, 0, 0, 1, 0, 1, 0]
y_pred = [1, 1, 0, 1, 1, 0, 1, 1, 0]
```

```python
cm = confusion_matrix(y_act, y_pred)
cm
```

```python
tn, fp, fn, tp = cm.ravel()
print('tn:', tn, 'fp:', fp, 'fn:', fn, 'tp:', tp)
```

```python
print('accuracy:', accuracy_score(y_act, y_pred))
print('precision:', precision_score(y_act, y_pred))
print('recall:', recall_score(y_act, y_pred))
print('specifity:', tn/(fp+tn))
print('f1:', f1_score(y_act, y_pred))
```

```python
plt.figure(figsize=(5, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Oranges", xticklabels=["predictedTrue", "predictedFalse"], yticklabels=["actualTrue", "actualFalse"])
plt.title("Actual vs Predicted")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
